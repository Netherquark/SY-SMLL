```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv("advertising.csv")
```

```python
df.head()
```

```python
x1 = df[["TV", "Radio", "Newspaper"]]
y = df["Sales"]
```

```python
reg = linear_model.LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(x1, y, test_size=0.2)
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_test["TV"], X_test["Radio"], y_test, color="blue")
ax.scatter(X_test["TV"], X_test["Radio"], y_pred, color="red")
plt.show() 
```

```python
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)
```

```python
df = pd.read_csv("Housing.csv")
```

```python
x1 = df[["lotsize", "bedrooms", "bathrms"]]
y = df["price"]
```

```python
reg = linear_model.LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(x1, y, test_size=0.2)
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
#plot the output
ax.scatter(X_test["lotsize"], X_test["bedrooms"], y_test, color="blue")
ax.scatter(X_test["lotsize"], X_test["bedrooms"], y_pred, color="red")
plt.show() # Show the plot
```

```python
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)
```
